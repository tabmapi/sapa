# Running SAPA without creating a new input file

Often, if you only want to make small changes to parameters in the input file, or e.g. change an element if you have some isostructural materials, you would have to make a new input file for SAPA. This can be cumbersome, so you can also run SAPA without creating a new input file, by editing a previous input file.

You can execute this input file in the following manner:

```python
import sys

sys.path.append('..')

import sapa as s

struc = s.sapa('BaTiO3_iso.cif')

temps = [15, 150, 210, 250, 293, 350, 410, 500]

struc.execute(temps)
```

When running in this manner, the execute function will ask you to supply a filename and sample name. The sample name is used to write the output file names. It can be found where the string 'sample' is in the below image - if you want to edit it, simply change the string in all 3 places (e.g. you could change 'sample' to 'sample2' or something to remind you of the changes you made to the input file, e.g. sample_finishx20 if you change the value of finish_x to 20)

Alternatively, you can specify the filename and sample before executing, as below:

```python
import sys

sys.path.append('..')

import sapa as s

struc = s.sapa('BaTiO3_iso.cif')

temps = [15, 150, 210, 250, 293, 350, 410, 500]

struc.sample = 'BaTiO3'

struc.filename = 'batch_modes'

struc.execute(temps)
```
